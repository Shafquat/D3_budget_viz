```python
import pandas as pd
```

```python
working_dir = "C:\\Users\\sarefee\\Documents\\R\\20-03-05 r2D3 Viz\\20-03-05 Budget Consolidation\\"
budget_2010_14 = "Budget_2010-2014.xlsx"
budget_2015 = pd.read_excel(working_dir + "approved-operating-budget-summary-2015.xlsx")
budget_2016 = pd.read_excel(working_dir + "approved-operating-budget-summary-2016.xlsx")
budget_2017 = pd.read_excel(working_dir + "approved-operating-budget-summary-2017.xlsx")
budget_2018 = pd.read_excel(working_dir + "approved-operating-budget-summary-2018.xlsx")
budget_2019 = pd.read_excel(working_dir + "approved-operating-budget-summary-2019_cleaned.xlsx")
```

```python
current_template = pd.read_excel(working_dir + budget_2010_14)
current_template
```

```python
def formater(budget,year):
    temp_df_exp = budget[budget['Expense/Revenue']=='Expenses'].groupby(['Program', 'Category Name']).sum().reset_index()
    temp_df_exp.rename(columns={temp_df_exp.columns[2]:'Expense','Category Name':'Category'}, inplace=True)
    temp_df_rev = budget[budget['Expense/Revenue']=='Revenues'].groupby(['Program', 'Category Name']).sum().reset_index()
    temp_df_rev.rename(columns={temp_df_rev.columns[2]:'Revenue','Category Name':'Category'}, inplace=True)
    temp_df_rev.Revenue = temp_df_rev.Revenue.abs()
    temp_df = pd.concat([temp_df_exp,temp_df_rev])
    temp_df['Year'] = year
    return temp_df
```

```python
b_15 = formater(budget_2015,2015)
b_16 = formater(budget_2016,2016)
b_17 = formater(budget_2017,2017)
b_18 = formater(budget_2018,2018)
b_19 = formater(budget_2019,2019)
```

```python
final_data = pd.concat([current_template,b_15,b_16,b_17,b_18,b_19])
final_data
```

```python
final_data.to_csv(working_dir + "budget_2010_2019.csv", index = False)
```
